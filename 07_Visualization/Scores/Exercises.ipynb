### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73], 
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

df = pd.DataFrame(data, columns = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])
df.head(5)
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(df.preTestScore, df.postTestScore, s=df.age)

plt.title("preTestScore and postTestScore")
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter(df.preTestScore, df.postTestScore, s= df.postTestScore * 4.5, c = df.female)

plt.title("preTestScore and postTestScore")
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

### BONUS: Create your own question and answer it.
Create a Bar group ofScores of preTest and postTest by name

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
df['name'] = df.first_name + ' ' + df.last_name

X = np.arange(len(df['name']))
ax.bar(X + 0.00, df['preTestScore'], color = 'r', width = 0.25)
ax.bar(X + 0.25, df['postTestScore'], color = 'c', width = 0.25)
ax.set_ylabel('Scores')
ax.set_title('Scores of preTest and postTest by name')
ax.set_xticks(X)
ax.set_xticklabels(df['name'])
ax.legend(labels=['preTestScore', 'postTestScore'])
plt.show()
```
