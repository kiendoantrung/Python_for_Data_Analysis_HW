# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
df = pd.read_csv(url)
df.head()
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python
plt.plot('month_number', 'total_profit', data=df)
plt.xticks(df['month_number'])
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title('Company profit per month')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
plt.plot('month_number', 'total_profit', data=df,  
         color='r', marker='o', linestyle='dashed', markerfacecolor = 'k', linewidth=3)

plt.xticks(df['month_number'])
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title('Company Sales data of last year')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(labels=['Profit data of last year'], loc = 'lower right')
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
plt.plot('month_number', 'facecream', data=df,  marker='o', linewidth=3)
plt.plot('month_number', 'facewash', data=df, marker='o', linewidth=3)
plt.plot('month_number', 'toothpaste', data=df, marker='o', linewidth=3)
plt.plot('month_number', 'bathingsoap', data=df, marker='o', linewidth=3)
plt.plot('month_number', 'shampoo', data=df, marker='o', linewidth=3)
plt.plot('month_number', 'moisturizer', data=df, marker='o', linewidth=3)

plt.title('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])

plt.legend(labels=["Face Cream Sales Data", "Face Wash Sales Data", "Toothpaste Sales Data",
                   "Bathingsoap Sales Data", "Shampoo Sales Data", "Moisturizer Sales Data"], loc='upper left')

plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python
plt.scatter('month_number', 'toothpaste', data=df)

plt.title(' Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.xticks(df['month_number'])
plt.legend(labels=['Toothpaste sales data'])
plt.grid(linestyle='--')
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
plt.bar(df['month_number'], df['facecream'], width= -0.35, align='edge')
plt.bar(df['month_number'], df['facewash'], width= 0.35, align='edge')

plt.title('Facewash and facecream sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.grid(linestyle="--")
plt.legend(labels = ['Face Cream sales data', 'Face Wash sales data'], loc='upper left')
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/06.png?raw=1)

```python
plt.bar(df['month_number'], df['bathingsoap'])

plt.title('bathing soap sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.grid(linestyle="--")
plt.show()

plt.savefig('bathing_soap_sales_data.png')
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/07.png?raw=1)

```python
range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(df['total_profit'], range)

plt.title('Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(labels = ['Profit data'],loc='upper left')
plt.xticks(range)
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/08.png?raw=1)

```python
categories = df.columns[1:7]
sales_data = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(),
              df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]

plt.axis("equal")
plt.pie(sales_data, labels=categories, autopct='%1.1f%%')
plt.title('Sales data')
plt.legend(loc='lower right')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/09.png?raw=1)

```python
fg, ax = plt.subplots(2, sharex=True)

ax[0].plot(df['month_number'], df['bathingsoap'], 
           label='Bathing sales data', c='k', marker='o', linewidth=3)
ax[0].set_title('Sales data of  a Bathingsoap')

ax[1].plot(df['month_number'], df['facewash'], 
           label = 'Face Wash Sales Data', c='r', marker='o', linewidth=3)
ax[1].set_title('Sales data of  a facewash')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/10.png?raw=1)

```python
plt.stackplot(df['month_number'], df['facecream'], df['facewash'], 
              df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer'], 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(labels=['face Cream', 'Face wash', 'Tooth paste', 
                   'Bating soap', 'Shampoo', 'Moisturizer'], loc='upper left')
plt.show()
```
